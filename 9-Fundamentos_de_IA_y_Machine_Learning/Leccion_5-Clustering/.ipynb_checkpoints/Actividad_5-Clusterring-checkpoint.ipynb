# Problema 1

Se pide agrupar un total de 8 patrones bidimensionales en tres clusters (k = 3).<br/>
   * Los patrones son los siguientes: 
       * A1 (2,10) A2 (2, 5)
       * A3 (8, 4) A4 (5, 8) 
       * A5 (7 ,5) A6 (6, 4)
       * A7 (1, 2) A8 (4, 9)
   * Los centroides iniciales son los puntos A1, A3 y A4.
   * La métrica de distancia utilizada será la distancia euclídea.

## ***1)* Representar los clusters creados y la posición de los centroides después de cada iteración.**

### Definamos la funciones necesarias

```python
def calc_dist(c1, c2, c3, df):
    dist = dict()
    dist[c1] = dict()
    C1 = np.array(c1)
    dist[c2] = dict()
    C2 = np.array(c2)
    dist[c3] = dict()
    C3 = np.array(c3)
    for item, row in df.iterrows():
      dist[c1][(row.X, row.Y)] = '{:.2f}'.format(np.linalg.norm(C1 - np.array(row)))
      dist[c2][(row.X, row.Y)] = '{:.2f}'.format(np.linalg.norm(C2 - np.array(row)))
      dist[c3][(row.X, row.Y)] = '{:.2f}'.format(np.linalg.norm(C3 - np.array(row)))      
    return dist
```

```python
def group(dist):
  k1 = list()
  D1 = dist[list(dist.keys())[0]]
  k2 = list()
  D2 = dist[list(dist.keys())[1]]
  k3 = list()
  D3 = dist[list(dist.keys())[2]]
  for item, row in df.iterrows():
    key = (row.X, row.Y)
    if D1[key] < D2[key] and D1[key] < D3[key]:
      k1.append((key, D1[key]))
    elif D2[key] < D3[key]:
      k2.append((key, D2[key]))
    else:
      k3.append((key, D3[key]))
  return [(list(dist.keys())[0], k1), (list(dist.keys())[1], k2), (list(dist.keys())[2], k3)]
```

```python
def calc_cent(groups):
  n_c = list()
  for group in groups:
    x, y = 0, 0
    for point in group[1]:
      x += float(point[0][0])
      y += float(point[0][1])
    x = float('{:.2f}'.format(x/len(group[1])))
    y = float('{:.2f}'.format(y/len(group[1])))
    n_c.append((x, y))
  return n_c
```

```python
def clustering(df, cents):
  k = 0
  clusters = dict()
  c01, c02, c03 = (0, 0), (0, 0), (0, 0)
  while c01 != cents[0] or c02 != cents[1] or c03 != cents[2]:
    loop = f'k_{k}'
    cluster = group(calc_dist(cents[0], cents[1], cents[2], df))
    clusters[loop] = cluster
    c01, c02, c03 = cents[0], cents[1], cents[2]
    cents = calc_cent(cluster)
    k += 1
  return clusters
```

```python
def draw_group(clusters, df):
  for c in clusters:
    fig, ax = plt.subplots()
    plt.scatter(df.X, df.Y)
    for k in clusters[c]:
      plt.scatter(k[0][0], k[0][1], marker='x')
      p_0 = 0.00
      for point in k[1]:
        if float(point[1]) > p_0:
          p_0 = float(point[1])
      circ = plt.Circle((k[0][0], k[0][1]), p_0 + 0.5, fill = False)
      ax.add_patch(circ)
    plt.show()
```

```python
def SSE(clusters):
  C_keys = list(clusters.keys())
  D = 0
  for c in clusters[C_keys[-1]]:
    for dist in c[1]:
      D += float(dist[1]) ** 2
  print(f'SSE = {D}')
```

### Empecemos los calculos:

```python
clusters = clustering(df, [(2, 10), (8, 4), (5, 8)])
draw_group(clusters, df)
```

## ***2)* El valor de la métrica SSE.**

```python
SSE(clusters)
```

# Problema 3


## Antes de nada, definamos las fucnciónes necesarias.

```python
def draw(e, df):
  fig, ax = plt.subplots()
  plt.scatter(df.X, df.Y)  # Dibujamos el scatter plot
  for item, row in df.iterrows():  # dibujamos los circulos:
    ax.add_patch(plt.Circle((row.X, row.Y), e, fill=False))
  plt.show()
```

```python
def exers(df, M, e):
  #adecuemos los datos al problema:
  points = list()
  for item, row in df.iterrows():
      points.append((row.X, row.Y))

  # mostremos la infromacion gráficamente:
  draw(e, df)

  # calculamos el modelo
  db = DBSCAN(eps=e, min_samples=M).fit(points)
  labels = db.labels_ # almacenamos la infromacion del entrenamiento
  colors = list()
  for l in labels:
    if l==1:
      colors.append('#37ff00')
    elif l==0:
      colors.append('#ff0000')
    else:
      colors.append('#000000')

  # Algo mas de informacion:
  no_clusters = len(np.unique(labels) )
  no_noise = np.sum(np.array(labels) == -1, axis=0)
  print(f'Número estimado de clusters: {no_clusters}') # analicemos la cantidad de clusters
  print('Incluyendo outliers')
  print(f'Númmero estimado de Outliers: {no_noise}') # y la cantidad de 'Outliers'

  # por ultimo imprimamaos por pantalla el resultado:
  plt.scatter(df.X, df.Y, c=colors)
  plt.show() 
```

Dados los mismos puntos del Problema I.

## ***1)* Aplicar DBSCAN considerando M = 3 y ε = √2.**

```python
exers(df, 3, np.sqrt(2))
```

## ***2)* Aplicar DBSCAN considerando M = 3 y ε = √10.**

```python
exers(df, 3, np.sqrt(10))
```

## ***3)* En que afecta el cvambio del parametro ε**

Dado que el parámetro ε establece el radio en el que debe presentarse un punto para que este se considere vecino de quien está siendo analizado, afectará al número de vecinos en torno a un punto por lo que afecta directamente a la generación de clúster. Si quisiéramos que los clústeres mostraran los puntos más estrechamente relacionados, estableceremos un ε menor, si la relación entre los puntos no necesitamos que sea tan estrecho, estableceremos un ε mayor. Esto afectará al número de clústeres directamente, por lo general a menor sea el ε, menor será el número de clústeres y viceversa.
