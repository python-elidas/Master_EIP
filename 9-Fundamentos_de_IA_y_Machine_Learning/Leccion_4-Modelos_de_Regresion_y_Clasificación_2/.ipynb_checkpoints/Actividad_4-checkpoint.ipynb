# Problema 1

***1)* Hallar las predicciones del modelo para el siguiente conjunto de test.**

Empecemos creando las funciones necesarias para la resolucion del ejercicio:

```python
# Empecemos importando las librerias necesarias:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import *
```

```python
# creemos la funcion para el calculo de Bn asumiendo que se trata de una red sigmoidal.
def B_n (bias, p1, p2, df, col_name):
    
    b = list()
    for item, row in df.iterrows():
        x = bias + (p1*row.X1) + (p2*row.X2)
        h = 1 / (1 + np.e ** (-x))
        b.append(float('{:.3f}'.format(h)))
    df[col_name] = b
    
```

```python
# creemos la funcion para el calculo de las predicciones
def p_Cn (bias, p1, p2, df, col_name):
    b = list()
    for item, row in df.iterrows():
        x = bias + (p1*row.B_1) + (p2*row.B_2)
        h = 1 / (1 + np.e ** (-x))
        b.append(float('{:.2f}'.format(h)))
    df[col_name] = b
    
```

```python
# definamos la funcion que según los datos obtenidos genere la prediccion:
def pred(df):
    
    pred = list()
    for item, row in df.iterrows():
        if row['p(C1)'] > row['p(C2)'] and row['p(C1)'] > row['p(C3)']:
            pred.append(1)
        elif row['p(C2)'] > row['p(C3)']:
            pred.append(2)
        else:
            pred.append(3)
    df['Y_p'] = pred
```

Apliquemos las funciones y tranformaciones:

```python
# leamos los datos:
df = pd.read_csv('Problema_1.csv', sep=';')
df
```

```python
# apliquemos la funcion refien creada:
B_n(6.46, -1.6, -3.43, df, 'B_1')
B_n(19.64, -1.35, -7.52, df, 'B_2')
y = df.Y
df = df.drop('Y', axis=1)
df['Y'] = y
df
```

```python
p_Cn(-5.47, 7.61, 1.9, df, 'p(C1)')
p_Cn(-3.76, -8.21, 7.87, df, 'p(C2)')
p_Cn(3.96, -3.81, -8.46, df, 'p(C3)')
y = df.Y
df = df.drop('Y', axis=1)
df['Y'] = y
df
```

```python
#apliquemos la funcion:
pred(df)
y = df.Y
df = df.drop('Y', axis=1)
df['Y'] = y
df
```

***2)* Evaluar el rendimiento del clasificador en dicho conjunto.**

```python
# Creamos la matriz de confusión:
cm = confusion_matrix(df.Y, df.Y_p)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
# Calculemos las variables para la clase 1
TP = np.diag(cm)[0]
FP = (cm.sum(axis=0) - TP)[0]  # columna (axis = 0 representa la suma por columnas)
FN = (cm.sum(axis=1) - TP)[0]  # fila (axis = 1 representa la suma por filas)
TN = np.diag(cm)[1] + np.diag(cm)[2]

# Precicion Global
ACC_1 = '{:.2f}'.format((TP+TN)/(TP+FP+FN+TN))
# Sensibilidad
TPR_1 = '{:.2f}'.format(TP/(TP+FN))
# False positive rate
FPR_1 = '{:.2f}'.format(FP/(FP+TN))
# Especificidad
TNR_1 = '{:.2f}'.format(TN/(TN+FP) )
# Precision
PPV_1 = '{:.2f}'.format(TP/(TP+FP))
#F1 score
F1_1 = '{:.2f}'.format((2*TP) / (2*TP+FP+FN))
```

```python
# Calculemos las variables para la clase 2
TP = np.diag(cm)[1]
FP = (cm.sum(axis=0) - TP)[1]  # columna (axis = 0 representa la suma por columnas)
FN = (cm.sum(axis=1) - TP)[1]  # fila (axis = 1 representa la suma por filas)
TN = np.diag(cm)[0] + np.diag(cm)[2]

# Precicion Global
ACC_2 = '{:.2f}'.format((TP+TN)/(TP+FP+FN+TN))
# Sensibilidad
TPR_2 = '{:.2f}'.format(TP/(TP+FN))
# False positive rate
FPR_2 = '{:.2f}'.format(FP/(FP+TN))
# Especificidad
TNR_2 = '{:.2f}'.format(TN/(TN+FP) )
# Precision
PPV_2 = '{:.2f}'.format(TP/(TP+FP))
#F1 score
F1_2 = '{:.2f}'.format((2*TP) / (2*TP+FP+FN))
```

```python
# Calculemos las variables para la clase 3
TP = np.diag(cm)[2]
FP = (cm.sum(axis=0) - TP)[2]  # columna (axis = 0 representa la suma por columnas)
FN = (cm.sum(axis=1) - TP)[2]  # fila (axis = 1 representa la suma por filas)
TN = np.diag(cm)[0] + np.diag(cm)[2]

# Precicion Global
ACC_3 = '{:.2f}'.format((TP+TN)/(TP+FP+FN+TN))
# Sensibilidad
TPR_3 = '{:.2f}'.format(TP/(TP+FN))
# False positive rate
FPR_3 = '{:.2f}'.format(FP/(FP+TN))
# Especificidad
TNR_3 = '{:.2f}'.format(TN/(TN+FP) )
# Precision
PPV_3 = '{:.2f}'.format(TP/(TP+FP))
#F1 score
F1_3 = '{:.2f}'.format((2*TP) / (2*TP+FP+FN))
```

```python
print('      | Precision Global | Sensibilidad | False Positive Rate | Especificidad | Precision | F1 Score')
print('------+------------------+--------------+---------------------+---------------+-----------+---------')
print(f'C1    |       {ACC_1}       |     {TPR_1}\t|\t{FPR_1}\t      |      {TNR_1}     |    {PPV_1}   |   {F1_1}')
print('------+------------------+--------------+---------------------+---------------+-----------+---------')
print(f'C2    |       {ACC_2}       |     {TPR_2}\t|\t{FPR_2}\t      |      {TNR_2}     |    {PPV_2}   |   {F1_2}')
print('------+------------------+--------------+---------------------+---------------+-----------+---------')
print(f'C3    |       {ACC_3}       |     {TPR_3}\t|\t{FPR_3}\t      |      {TNR_3}     |    {PPV_3}   |   {F1_3}')  
```

### Conclusion:

Como bien se puede observar en la tabla mostrada, el ratio de falsos positivos es bastante alto y la sensibilidad bastante baja. Ello unido al hecho de que la especificidad también es elevada y que la precisión en ningún caso supera el 80%, podemos decir que el modelo no es especialmente preciso.
