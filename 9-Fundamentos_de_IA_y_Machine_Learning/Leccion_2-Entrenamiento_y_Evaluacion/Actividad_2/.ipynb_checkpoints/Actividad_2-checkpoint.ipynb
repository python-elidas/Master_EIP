```python
# Importemos las librerias necesarias:
import pandas as pd
import numpy as np
from sklearn.metrics import *
import matplotlib.pyplot as plt
```

# Ejercicio 1:
**Dado un conjunto de 30 datos de test, con la variable objetivo real y la
salida proporcionada por dos modelos.**

*a)* Calcular las métricas de regresión para cada modelo.

```python
# Empecemos cargando el CSV con los datos:
df = pd.read_csv('Ejercicio_1.csv', sep=';')
for row in df:
    df[row] = pd.to_numeric(df[row])
df
```

```python
print(f'   Error Absoluto Médio | Raiz del Error Cuadrático Médio | Coeficiente de deterrminacion')
print(f'M1  {mean_absolute_error(df.Y, df.M1)}  |\t{mean_squared_error(df.Y, df.M1)}\t  |\t{r2_score(df.Y, df.M1)}')
print(f'M2  {mean_absolute_error(df.Y, df.M2)}  |\t{mean_squared_error(df.Y, df.M2)}\t  |\t{r2_score(df.Y, df.M2)}')
```

*b)* En función de los resultados, decidir qué modelo es mejor.

Una vez obtenidos los datos de los errores de los modelos, podemos establecer en función de los datos la precisión de los modelos.</br>
Como bien se puede observar, los valores del Error Absoluto Medio y de la Raíz del Error Cuadrático Medio son menores para el modelo M2. Del mismo modo el Coeficiente de determinación es más cercano a 1 para este modelo por lo que se puede determinar que el modelo M2 es más preciso que el modelo M1.

# Ejercicio 2:

**Dado un problema de clasificación binaria (sólo existen dos clases),
tenemos la salida de cada uno de 30 patrones en test para dos
modelos distintos.**

*a)* Montar la matriz de confusión de cada modelo.

```python
# Empecemos Creando el data frame:
# Empecemos cargando el CSV con los datos:
df = pd.read_csv('Ejercicio_2.csv', sep=';')
for row in df:
    df[row] = df[row].astype(int)
df
```

```python
#calculemos la matriz de confucion para el modelo M1:
cm_M1 = confusion_matrix(df.Y, df.M1)
M1_disp = ConfusionMatrixDisplay(confusion_matrix=cm_M1)
M1_disp.plot()
plt.show()
```

```python
#calculemos la matriz de confucion para el modelo M1:
cm_M2 = confusion_matrix(df.Y, df.M2)
M2_disp = ConfusionMatrixDisplay(confusion_matrix=cm_M2)
M2_disp.plot()
plt.show()
```

*b)* Calcular las métricas de clasificación, excepto el AUC.

```python
# Calculemos las variables necesarias para el modelo M1
TP_1 = np.diag(cm_M1)[0]
FP_1 = (cm_M1.sum(axis=0) - TP_1)[0]
FN_1 = (cm_M1.sum(axis=1) - TP_1)[0]
TN_1 = cm_M1.sum() - (FP_1 + FN_1 + TP_1)
print(TP_1, FP_1, FN_1, TN_1)

# Precicion Global
ACC_1 = (TP_1+TN_1)/(TP_1+FP_1+FN_1+TN_1)
# Sensibilidad
TPR_1 = TP_1/(TP_1+FN_1)
# False positive rate
FPR_1 = FP_1/(FP_1+TN_1)
# Especificidad
TNR_1 = TN_1/(TN_1+FP_1) 
# Precision
PPV_1 = TP_1/(TP_1+FP_1)
#F1 score
F1_1 = f1_score(df.Y, df.M1)
```

```python
# Calculemos las variables necesarias para el modelo M2
TP_2 = np.diag(cm_M2)[0]
FP_2 = (cm_M2.sum(axis=0) - TP_2)[0]
FN_2 = (cm_M2.sum(axis=1) - TP_2)[0]
TN_2 = cm_M2.sum() - (FP_2 + FN_2 + TP_2)

# Precicion Global
ACC_2 = (TP_2+TN_2)/(TP_2+FP_2+FN_2+TN_2)
# Sensibilidad
TPR_2 = TP_2/(TP_2+FN_2)
# False positive rate
FPR_2 = FP_2/(FP_2+TN_2)
# Especificidad
TNR_2 = TN_2/(TN_2+FP_2) 
# Precision
PPV_2 = TP_2/(TP_2+FP_2)
#F1 score
F1_2 = f1_score(df.Y, df.M2)
```

```python
# Imprimamos los resaultados
print('   Precision Global   | Sensibilidad | False Positive Rate | Especificidad | Precision\t\t| F1 Score')
print(f'M1 {ACC_1} |     {TPR_1}     |\t\t{FPR_1}\t   |\t{TNR_1}\t   | {PPV_1} | {F1_1}')
print(f'M2 \t{ACC_2}\t      |     {TPR_2}      |\t\t{FPR_2}\t   |\t{TNR_2}\t   | {PPV_2} | {F1_2}')
```

En función de los datos obtenidos y si observamos el ratio de falsos positivos, la especificidad y la precisión de los modelos, podemos concluir que el mejor modelo es el M2.

# Ejercicio 3

**Dada la siguiente matriz de confusión para un problema multiclase, se
pide hallar todas las métricas de clasificación para cada clase, excepto
el AUC. Recuerde que, para calcular las métricas de una clase, se
considera dicha clase como positiva y el resto como la negativa.**

```python
# Primero crearemos la matriz en forma de lista de listas:
cm = [[20, 10, 5], [5, 30, 0], [5, 5, 25]]
```

```python
# Creemos ahora las variables necesarias para la clase GATO:
TP, TN, FP, FN = 20, 55, 10, 15
# Precicion Global
ACC = (TP + TN) / (TP + FP + FN + TN)
# Sensibilidad
TPR = TP / (TP + FN)
# False positive rate
FPR = FP / (FP + TN)
# Especificidad
TNR = TN / (TN + FP) 
# Precision
PPV = TP / (TP + FP)
#F1 score
F1 = (2 * TP) / ((2 * TP) + FP + FN) 
print(f'Métricas de Clasificación para la clase gato:\n\
    Precision Global:    {ACC}\n\
    Sensibilidad:        {TPR}\n\
    False positive rate: {FPR}\n\
    Especificidad:       {TNR}\n\
    Precision:           {PPV}\n\
    F1 score:            {F1}')
```
