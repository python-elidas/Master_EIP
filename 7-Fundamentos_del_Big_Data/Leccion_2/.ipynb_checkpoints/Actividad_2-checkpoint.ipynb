___Segunda parte de la Actividad de la Lección 2___

Ggplot es una librería muy importante en “R” (lenguaje de programación).
En el caso de Python, lo que se va a pedir es que se aprenda de forma
autodidacta a hacer las gráficas para Iris Dataset con “plotnine”.

Segun la infomacion hallada, ggplot2 es una de las librerias mas empleadas en R para el manejo de graficas ya que ofrece gran cantidad de opciones y permite modificar practicamente cada elemento que compone una grafica. Por la parte de pytohn teneoms la librería plitnine.

```python
df = pd.read_csv('Iris_DataSheet.csv')
df.dropna() # eliminamos los valores (nulos en caso de haberlos)
```

```python
column_names = list(df.columns) #obtenemos los nombres de las columnas
column_names
```

```python
#creamos el grafico para el Sépalo:
sepal = (
    gg.ggplot(df)
    + gg.aes(x=column_names[1], y=column_names[2], color=column_names[-1])
    + gg.geom_point(size=1)
    + gg.ggtitle('Iris Data Sheet by Sepal')
)
sepal
```

```python
#creamos el grafico para el Petalo:
petal = (
    gg.ggplot(df)
    + gg.aes(x=column_names[3], y=column_names[4], color=column_names[-1])
    + gg.geom_point(size=1)
    + gg.ggtitle('Iris Data Sheet by Petal')
)
petal
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
```

```python
colors = []
for elem in df[column_names[-1]]:
    if elem == 'Iris-setosa':
        colors.append('red')
    elif elem == 'Iris-versicolor':
        colors.append('blue')
    elif elem == 'Iris-virginica':
        colors.append('green')
df['color'] = colors #creamos una columna con los colores que le corresponde a cada especie
column_names = list(df.columns) # actualizamos la lista de los nombres de las columnas
```

```python
# graficamos según sépalo
sepal_plot = figure()
sepal_plot.circle(df[column_names[1]], df[column_names[2]], fill_color=df[column_names[-1]])
show(sepal_plot)
```

```python
#graficamos según petalo
sepal_plot = figure()
sepal_plot.circle(df[column_names[3]], df[column_names[4]], fill_color=df[column_names[-1]])
show(sepal_plot)
```
