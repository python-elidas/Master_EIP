_-Pregunta 2 – (hasta 1 punto)_<br/>
¿ Qué son Q1, Q2, Q3 ?

Son los denominados cuartiles, cada uno de ellos delimita un cuarto de la poblacion, dicho de otro modo, el valor Q1 es aquel para el cual el 25% de los datos permanecen por debajo de este, Q2 es el valor para el cual el 50% de los datos pernamecen por debajo de el, también se le denomina Media y por último, Q3 es el valor para el cual el 75% de los datos permanecen por debajo de el.

_-Pregunta 3 – (hasta 1 punto)_<br/>
¿ Qué son los “outliers” ? ¿ son mostrados en estos diagramas de cajas?

Los 'Outliers' o valores atipicos son aquellos valores que se salen de la "Normalidad" de los datos, empleando el ejemplo del Titanic DataSheet, si observamos el BoxPlot generado en la primera pregunta, se observan una serie de rombos por encima (podrian tambein estar por debajo) de los boxplot, estos son los outlier.
