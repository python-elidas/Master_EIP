```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = [8,1,5,2,4]
y = ['Aplicaciones', 'Fund BigData', 'IA', 'ScrumMaster', 'Pentesting']
df = pd.DataFrame({'X': y,'y': X})
#NombreDf.NombreColumna
df.y.plot(kind='bar')
plt.show()
#__Con esto se crea y se muestra un grafico__
```

```python
df.y.value_counts()
#__con esto obtenemos la cantidad de veces que aparece cada valor.
```

__HISTOGRAMA__

```python
plt.hist(df.y)
plt.grid(True)
plt.xlabel('X_Label')
plt.ylabel('y_Label')
plt.title('Histograma')
plt.show()
#__Con esto se crea y se configura un histograma
```

__SCATTER PLOT__

```python
plt.scatter(df.X,df.y)
plt.grid(True)
plt.xlabel('X_Label')
plt.ylabel('y_Label')
plt.title('Scatter_Plot')
plt.show()
#Con esto se crea y se configura un scatter_plot o grafico de puntos
```

__PIE_CHART__

```python
X = [8,1,5,2,4]
y = ['Aplicaciones', 'Fund BigData', 'IA', 'ScrumMaster', 'Pentesting']
```

```python
plt.title('Grafico Circular')
plt.pie(X, labels=y)
plt.show()
```
