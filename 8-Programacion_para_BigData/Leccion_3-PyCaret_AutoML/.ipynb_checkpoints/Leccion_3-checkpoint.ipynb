## Importaci√≥n de dependencias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importaremos todo pycaret Classificacion
from pycaret.classification import *
```

## Importamos el Titanic Data Set

```python
df = pd.read_csv("Titanic_DS/train.csv")
df.head()
```

## Clasificamos las columnas relevantes 

```python
clf = setup(
    data = df, # de donde sacamos los datos
    target = 'Survived', # columna de referencia
    ignore_features = ['Name', 'Ticket', 'PassengerId'] # columnas que no son relevantes para la predicci√≥n
    )

# Como se observa en este caso, no es necesario modificar el data set para eliminiar las columnas que no son 
# releventes, vasta con decrile que las ignore.
```

## compare_models()

```python
# con esta funcion se van a comparar los diferentes modelos y nos dar√° la precisi√≥n y otra serie de
# datos que ser√°n relevantes a la hora de elegir un modelo u otro.

compare_models()
```

## create_model()

```python
# Ahora procederemos a crear el modelo fijandonos en los valores obtenidos en la l√≠nea anterior.
# Si bien es cierto que se han obtenido bastantes valores, la mayoria se exlpicar√°n en la asignatura de
# Machine Learning. De momento, nos quedaremos con los parametros "Acuracy" y "AUC" de modo que el modelo a emplear
# ser√° "gbc" o "Gradien Boosting Classifier".

gbc = create_model('gbc')
```

Crearemos tambien otros dos modelos con los siguientes modelos mas precisoso a fin de tener algo con lo que comparar

```python
#Random Forest
rf = create_model('rf')
```

```python
#Logistic Regression
lr = create_model('lr')
```

## tune_model

Pasaremos ahora a "ecualizar" los modelos, podremos pedir (de momento) que realice la optimizaci√≥n por 
"Accuracy" o por "AUC" (Area Under the Curbe)

```python
# gbc optimizado mediante Accuracy
tune_gbc_acc = tune_model(gbc, optimize='Accuracy')
```

```python
# gbc optimizado mediante AUC
tune_gbc_auc = tune_model(gbc, optimize='AUC')
```

```python
# lr optimizado mediante Accuracy
tune_lr_acc = tune_model(lr, optimize='Accuracy')
```

```python
# lr optimizado mediante AUC
tune_lr_auc = tune_model(lr, optimize='AUC')
```

```python
# ridge optimizado mediante Accuracy
tune_rf_acc = tune_model(rf, optimize='Accuracy')
```

```python
# ridge optimizado mediante AUC
tune_rf_auc = tune_model(rf, optimize='AUC')
```

### Mostremos unos graficos:

Emplearemos el modelo que mejor resultado haya dado, en mi caso es el mdelo gbc previo a ser optimizado.

existen varios graficos con los que se puede evaluar la precisi√≥n del modelo sin embargo, existe una forma de poder verlos todos:
_evaluate_model()_

```python
evaluate_model(gbc)
```

```python
evaluate_model(tune_gbc_acc)
```

```python
evaluate_model(tune_gbc_auc)
```

Como se puede observar en los gracifos mostrados, sobre todo en Learning Curve y Validation Curve, una vez los modelos han sido optimizados hacen cosas raras por lo que emplearemos el modelo gbc sin optimizaci√≥n alguna.

Creamos el archivo CSV para subirlo a Kaggle

### Hemos usado la Confusion Matrix en esta lecci√≥n, pero no hemos explicado mucho, haz un peque√±o resumen con informaci√≥n que encuentres en Internet acerca de esto.

La "confusion matrix" representa de forma visual y sencilla la precisi√≥n del modelo empleado. En las columnas se representan los valores predichos y en las filas los reales, de modo que se establecen cuatro grupos, Positivo Real (predicci√≥n positiva correcta), Negativo Real (predicci√≥n Negativa Correcta), Falso Positivo (predicci√≥n Positiva Incorrecta) y Falso Negativo(predicci√≥n Negativa Incorrecta). Este tipo de matrices est√°n directamente relacionadas con las Curvas AUC-ROC.
