## Importación de dependencias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importaremos todo pycaret Classificacion
from pycaret.classification import *
```

## Importamos el Titanic Data Set

```python
df = pd.read_csv("Titanic_DS/train.csv")
df.head()
```

## Clasificamos las columnas relevantes 

```python
clf = setup(
    data = df, # de donde sacamos los datos
    target = 'Survived', # columna de referencia
    ignore_features = ['Name', 'Ticket', 'PassengerId'] # columnas que no son relevantes para la predicción
    )

# Como se observa en este caso, no es necesario modificar el data set para eliminiar las columnas que no son 
# releventes, vasta con decrile que las ignore.
```

## compare_models()

```python
# con esta funcion se van a comparar los diferentes modelos y nos dará la precisión y otra serie de
# datos que serán relevantes a la hora de elegir un modelo u otro.

compare_models()
```

## create_model()

```python
# Ahora procederemos a crear el modelo fijandonos en los valores obtenidos en la línea anterior.
# Si bien es cierto que se han obtenido bastantes valores, la mayoria se exlpicarán en la asignatura de
# Machine Learning. De momento, nos quedaremos con los parametros "Acuracy" y "AUC" de modo que el modelo a emplear
# será "gbc" o "Gradien Boosting Classifier".

gbc = create_model('gbc')
```

Crearemos tambien otros dos modelos con los siguientes modelos mas precisoso a fin de tener algo con lo que comparar

```python
#Random Forest
rf = create_model('rf')
```

```python
#Logistic Regression
lr = create_model('lr')
```

## tune_model

Pasaremos ahora a "ecualizar" los modelos, podremos pedir (de momento) que realice la optimización por 
"Accuracy" o por "AUC" (Area Under the Curbe)

```python
# gbc optimizado mediante Accuracy
tune_gbc_acc = tune_model(gbc, optimize='Accuracy')
```

```python
# gbc optimizado mediante AUC
tune_gbc_auc = tune_model(gbc, optimize='AUC')
```

```python
# lr optimizado mediante Accuracy
tune_lr_acc = tune_model(lr, optimize='Accuracy')
```

```python
# lr optimizado mediante AUC
tune_lr_auc = tune_model(lr, optimize='AUC')
```

```python
# ridge optimizado mediante Accuracy
tune_rf_acc = tune_model(rf, optimize='Accuracy')
```

```python
# ridge optimizado mediante AUC
tune_rf_auc = tune_model(rf, optimize='AUC')
```

### Mostremos unos graficos:

Emplearemos el modelo que mejor resultado haya dado, en mi caso es el mdelo gbc previo a ser optimizado.

existen varios graficos con los que se puede evaluar la precisión del modelo sin embargo, existe una forma de poder verlos todos:
_evaluate_model()_

```python
evaluate_model(gbc)
```

```python
evaluate_model(tune_gbc_acc)
```

```python
evaluate_model(tune_gbc_auc)
```

Como se puede observar en los gracifos mostrados, sobre todo en Learning Curve y Validation Curve, una vez los modelos han sido optimizados hacen cosas raras por lo que emplearemos el modelo gbc sin optimización alguna. 

## Empezamos con test.csv

```python
test = pd.read_csv('Titanic_DS/test.csv')
test.head()
```

Creamos las predicciones con el modelo que mejor resultado haya dado hasta el momento:

```python
pred = predict_model(gbc, data = test)
pred
```

Separamos la infromación que nos interesa:

```python
pred_sub = pred[['PassengerId', 'Label']]
pred_sub.head()
```

Abrimos el gender_submision.csv

```python
sub = pd.read_csv('Titanic_DS/gender_submission.csv')
sub.head()
```

Modificamos la información para que figure nuestra predicción

```python
sub['PassengerId'] = pred_sub['PassengerId']
sub['Survived'] = pred_sub['Label']
sub.head()
```

Creamos el archivo CSV para subirlo a Kaggle

```python
# guardaremos el archivo en una nueva carpeta dentro de Titanic_DS llamada Predicciones
sub.to_csv('Titanic_DS/predicciones/titanic_1.csv', index = False)
```

### Hemos usado la Confusion Matrix en esta lección, pero no hemos explicado mucho, haz un pequeño resumen con información que encuentres en Internet acerca de esto.

La "confusion matrix" representa de forma visual y sencilla la precisión del modelo empleado. En las columnas se representan los valores predichos y en las filas los reales, de modo que se establecen cuatro grupos, Positivo Real (predicción positiva correcta), Negativo Real (predicción Negativa Correcta), Falso Positivo (predicción Positiva Incorrecta) y Falso Negativo(predicción Negativa Incorrecta). Este tipo de matrices están directamente relacionadas con las Curvas AUC-ROC.
